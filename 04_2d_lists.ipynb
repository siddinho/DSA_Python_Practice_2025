# 2D Lists

```python
l = [[1,2,3],[4,5,6],[7,8,9]]
print(l)
```

```python
nrows= len(l)
ncols = len(l[0])
```

# Print a 2d lists

```python
for i in range(nrows):
    for j in range(ncols):
        print(l[i][j],end = " ")
    print()
```

# Q - Print upper triangle - output - <br>
1 2 3 <br>
% 5 6 <br>
% % 9

```python
for i in range(nrows):
    stars = 0
    while stars<i:
        print("*",end = " ")
        stars = stars + 1
    for j in range(stars,ncols):
        print(l[i][j],end = " ")
    print()
```

## Easy solution

```python
for i in range(nrows):
    for j in range(ncols):
        if j>=i:
            print(l[i][j],end = " ")
        else:
            print("*",end = " ")
    print("")
        
```

# Q - Print Lower Triangle
1 * * <br>
4 5 * <br>
7 8 9 <br>
logic - if j<=i then print number else print "*"

```python
for i in range(nrows):
    for j in range(ncols):
        if j<=i:
            print(l[i][j],end = " ")
        else:
            print("*",end = " ")
    print("")
        
```

# Q -  Print Diagonal <br>
1 % % <br>
% 5 % <br>
% % 9 <br>

logic = only print if i ==j  else "%"

```python
for i in range(nrows):
    for j in range(ncols):
        if j==i:
            print(l[i][j],end = " ")
        else:
            print("*",end = " ")
    print("")
        
```

## Print Diagonal D2

output <br>
% % 3 <br>
% 5 % <br>
7 % % <br>  

```python
for i in range(nrows):
    for j in range(ncols):
        if i+j==2:
            print(l[i][j],end = " ")
        else:
            print("*",end = " ")
    print("")
        
```

# Transpose of a Matrix

```python
l = [[1,2,3],[4,5,6]]

nrows = len(l)
ncols = len(l[0])
output_array = [[0]*nrows for _ in range(ncols)]
print(output_array)

for i in range(nrows):
    for j in range(ncols):
        output_array[j][i] = l[i][j]

print("Output")
print(output_array)
```
