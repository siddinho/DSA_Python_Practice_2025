# Sorting

## Selection Sort - 
Approach - Start from i till n: <br>
assume i to be min_index <br>
for j in i+1,n: <br>
------->get the min index - if arr[j]<arr[min_index] then min_index = j <br>
swap(arr[i],arr[min_index])

<hr>
selecting min index in each pass and swapping

```python
l = [5,2,3,4,1,10,15,11]
## own code
def ascending_selection_sort(l:list):
    if len(l)<=1:
        return l
    else:
        for i in range(len(l)-1):
            min_index = i
            for j in range(i+1,len(l)):
                if l[j]<l[min_index]:
                    min_index = j
            l[i],l[min_index] = l[min_index],l[i]
        return l
```

```python
ascending_selection_sort(l)
```

```python
def descending_selection_sort(arr):
    n = len(arr)
    for i in range(n):
        max_index = i
        for j in range(i+1,n):
            if arr[j]>arr[max_index]:
                max_index = j
        arr[i],arr[max_index] = arr[max_index],arr[i]
    return arr
```

```python
arr = [3,1,4,2]
arr= [0]
descending_selection_sort(arr)
```

# Bubble Sort
Approach adjacent swaps , largest element comes in last in ascending order

```python
arr = [3,1,2,3,4,-1,-2,-3]
def bubble_sort(arr):
    n= len(arr)
    for i in range(1,n):
        for j in range(n-i):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]

    return arr
```

```python
bubble_sort(arr)
```

# Insertion Sort 

Keep checking if the element is at right place in the subarray moving fom left to right,
if not then move the j and j +1 and insert in the subarray <br>
## Move array then insert only once

```python
def insertion_sort(arr): # Done with help
    n = len(arr)

    for i in range(1, n):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]   # shift right
            j -= 1

        arr[j + 1] = key          # insert key
    return arr
```

```python
arr = [1,2,4,3,-1]
insertion_sort(arr)
```

# Merge Sort

## Merge 2 sorted arrays

```python
def merge_2_sorted_arrays(left_arr,right_arr):
    i = 0
    j = 0
    left_length = len(left_arr)
    right_length = len(right_arr)
    res = []

    while i<left_length and j<right_length: # Common elements
        if left_arr[i]<=right_arr[j]:
            res.append(left_arr[i])
            i = i +1
        else:
            res.append(right_arr[j])
            j = j+1
    if i ==left_length and j!=right_length: # left is exhausted
        res.extend(right_arr[j:])
    else:
        res.extend(left_arr[i:])
    return res
    
```

```python
left = [1,2,3,4,5]
right = [-1,4,5,6]
merge_2_sorted_arrays(left_arr=left,right_arr=right)
```

## Merge Sort - Recursive Function

```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    else:
        n = len(arr)
        mid = n//2
        left_arr = arr[:mid]
        right_arr = arr[mid:]
        left_sort = merge_sort(left_arr)
        right_sort = merge_sort(right_arr)
        res = merge_2_sorted_arrays(left_arr=left_sort,right_arr=right_sort)
        return res
```

```python
arr = [4,5,6,7,8,1,2,3,4,5,10,11,-1]
merge_sort(arr)
```

# Quick Sort

```python
def partition(arr,low,high):
    n = len(arr)
    p = arr[low]
    i = low
    j = high

    while i <j:
        while arr[i]<=p and i<=high-1:
            i = i+1
        while arr[j]>p and j>=low+1:
            j = j -1
        if i<j:
            arr[i],arr[j] = arr[j],arr[i]
    arr[low],arr[j] = arr[j],arr[low]
    return j

def quick_sort(arr,low,high):
    if low<high:
        p_index = partition(arr,low,high)
        quick_sort(arr,low,p_index-1)
        quick_sort(arr,p_index+1,high)

```

```python
arr = [4,1,2,3,5,6]
quick_sort(arr,0,len(arr)-1)
```
