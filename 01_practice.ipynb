# Palindrome - Number

```python
n = 1

num = n
rd = 0

while num >0:
    rem = num % 10
    rd = rd*10 + rem
    num = num // 10
print(rd)
if rd ==n:
    print(True)
else:
    print(False)
```

# Armstrong Number
n = 123
then 1^3 + 2^3 + 3^3 = 123 Armstrong number

```python
from math import *

n  = 1534
number_of_digits = int(log10(n))+1
print(number_of_digits)
num = n
sum = 0

while num >0:
    rem = num%10
    sum = sum + (rem**number_of_digits)
    print(rem,sum)
    num = num//10
print(sum)
sum == n



```

# Print Factors of a number

```python
num = 6 ## Brute Force
i = 1
while i <= num:
    if num % i ==0:
        print(i)
    i = i +1
```

```python
for i in range(1,num//2+1,1): ## O(N/2)
    if num%i == 0:
        print(i)
print(num)
```

## Best Solution

```python
import numpy as np
n = 36
fac = []
end = int(np.sqrt(n))

for i in range(1,end+1):
    if n%i == 0:
        fac.append(i)
        dig = n // i
        if dig != i:
            fac.append(dig)
```

## Frequency Mapping/ hashing

```python
l = [1,222222,3,1,2,5,3]
d = {}

for i in l:

    if d.get(i) is None:
        d[i] = 1
    else:
        d[i] = d[i]+1
```

# Recursion - Function that calls itself,
needs to have a base case from which it returns

## Q - Print Siddharth 6 times using Recursion

## Head Recursion - Processing first then function call

```python
count = 0

def func():
    global count
    if count == 4:
        return
    print("Sid {0}".format(count))
    count = count + 1
    func() ## Recusrive Call post function call

func()
```

## Tail Recursion - Back tracking

```python
count = 0

def func_1():
    global count
    if count == 4:
        return
    count = count + 1
    func_1() 
    print("Sid {0}".format(count))

func_1()
```

## Recursion using parameters

### Q - Print x, N times using recursion

```python
def p(x,n): ## Own approach
    if n ==0:
        return
    print(x)
    p(x,n-1)
```

```python
p(1,5)
```

### Q - Print 1 to n using recursion

```python
def fn(x,n): ## Head Recursion
    if x ==n+1:
        return
    print(x)
    x = x + 1
    fn(x,n)
```

```python
fn(1,6)
```

```python
def fn_1(n): ## Tail Recursion for printing 1 to N
    if n==0:
        return
    fn_1(n-1)
    print(n)
```

```python
fn_1(5)
```
