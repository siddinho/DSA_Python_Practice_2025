## Best Solution

```python
import numpy as np
n = 36
fac = []
end = int(np.sqrt(n))

for i in range(1,end+1):
    if n%i == 0:
        fac.append(i)
        dig = n // i
        if dig != i:
            fac.append(dig)
```

# Recursion - Function that calls itself,
needs to have a base case from which it returns

## Q - Print Siddharth 6 times using Recursion

## Head Recursion - Processing first then function call

```python
count = 0

def func():
    global count
    if count == 4:
        return
    print("Sid {0}".format(count))
    count = count + 1
    func() ## Recusrive Call post function call

func()
```

## Tail Recursion - Back tracking

```python
count = 0

def func_1():
    global count
    if count == 4:
        return
    count = count + 1
    func_1() 
    print("Sid {0}".format(count))

func_1()
```

## Recursion using parameters

### Q - Print x, N times using recursion

```python
def p(x,n): ## Own approach
    if n ==0:
        return
    print(x)
    p(x,n-1)
```

```python
p(1,5)
```

### Q - Print 1 to n using recursion

```python
def fn(x,n): ## Head Recursion
    if x ==n+1:
        return
    print(x)
    x = x + 1
    fn(x,n)
```

```python
fn(1,6)
```

```python
def fn_1(n): ## Tail Recursion for printing 1 to N
    if n==0:
        return
    fn_1(n-1)
    print(n)
```

```python
fn_1(5)
```

## Paramterized Recursion

### Q -  Sum of n numbers

```python
def add(sum,i,n):
    if i >n:
        print(sum)
        return
    add(sum+i,i+1,n)
```

## Functional Recursion

```python
def add_1(n): ## Done on my own
    if n == 1:
        return 1
    sum = n + add_1(n-1)
    return sum
```

```python
add_1(10)
```

### Factorial

```python
def fac(n): ## Done on my own
    if n ==0:
        return 1
    return n*fac(n-1)

fac(0)
```

## Reverse array between indexes using Recursion

```python
def rev(arr,left,right):
    if left == right:
        return 
    t = arr[left]
    arr[left] = arr[right]
    arr[right] = t
    rev(arr,left+1,right-1)
```

```python
arr = [1,2,3,4,5]
left = 0
right = 4

rev(arr,left,right)
```

```python
arr
```

# String Palindrome or not

## Brute Force - Naive

```python
s = "nitinaa" ## Reversing full string and checking

if s == s[::-1]:
    print(True)
else:
    print(False)
```

## Optimized while - step by step checking -  double pointer

```python
s = "nitin"
left  = 0
right = len(s)-1
flag = True
while left < right:
    if s[left]!=s[right]:
        flag = False
        break
    left+=1
    right-=1
print(flag)
```

## Recursion

```python
def check_palindrome(s,left,right): ## Done on own
    if left >= right:
        return True
    if s[left]!=s[right]:
        return False
    return   check_palindrome(s,left+1,right-1)
s = "abccba"
check_palindrome(s,0,len(s)-1)
```

# Find nth Fibonacci Number

```python
n = 9

count = 1
first = 0
last = 1
sum =1

while count <n:
    sum = first+last
    first = last
    last = sum
    count = count +1

print(sum)
```

## Using Recursion

```python
def fib(n):
    if n ==0:
        return 0
    if n == 1:
        return 1
    return (fib(n-1)+fib(n-2))

fib(5)
```

# Reverse an array using recursion

```python
def rev_rec(arr,left,right):
    if left>=right:
        return arr
    arr[left],arr[right] = arr[right],arr[left]
    rev_rec(arr,left+1,right-1)


arr = [1,2,3,4,5]
rev_rec(arr,0,len(arr)-1)
```
