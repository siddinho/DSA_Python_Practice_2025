### Q3 - Check if array is sorted or not

```python
def check_array(arr):  ## Own Approach
    n = len(arr)
    asc = True

    if n <= 1:
        return True,asc
    elif n == 2:
        diff = arr[1]-arr[0]
        if diff>=0:
            return True,asc
        else:
            asc = False
            return True,asc
    else:
        diff = arr[1]-arr[0]
        if diff>=0:
            sort_p = True
            sort_n = False
        else:
            sort_n = True
            sort_p = False
            asc = False
        for i in range(n-1):
            diff = arr[i+1]-arr[i]
            if sort_p and diff<0:
                return False,False
            elif sort_n and diff>0:
                return False,False
            
    return True,asc

        
    
```

### Q4 Remove duplicates from a sorted array in place
input = [1,1,1,1,1,2,2,2,3,3,4] <br>
output = [1,2,3,4,-1,-1,-1,-1,-1]

```python
def remove_duplicates_sorted_arr(arr):
    if not arr:
        return 0

    write = 1

    for read in range(1, len(arr)):
        if arr[read] != arr[read - 1]:
            arr[write] = arr[read]
            write += 1

    return write  # length of unique array
```

```python
arr = [1,1,1,1,2,2,2,3,4,5]
k = remove_duplicates_sorted_arr(arr)
arr[0:k]
```

```python
def remove_dup_array(arr):
    n = len(arr)
    if n ==1:
        return 1
    i = 0
    j = 1
    while j <n:
        if arr[i]!=arr[j]:
            i = i +1
            arr[i],arr[j] = arr[j],arr[i]
        j= j +1
    return i+1 ## Unique elements
```

```python
arr  = [11,1,1,1,2,2,2,4,4,5,6]
k1 = remove_dup_array(arr)
arr[0:k1]
```

```python
def dup_2(arr):
    n = len(arr)
    first = 0
    second = 1
    k = 0

    while second < n:
        if arr[first]!=arr[second]:
            arr[first+1] = arr[second]
            first = first+1
            k = k +1
        second = second +1
    return arr[0:k+1]
```

```python
arr = [1,1,1,1,2,2,2,3,4,5,5,5,5,8,9,9,10]
arr=[0,1]
dup_2(arr)
```

# Rotate Right by k places

```python
def right_rotate_k(arr,k):
    n = len(arr)
    k = k%n ## Effective k 
    temp = arr[n-k:n]

    for i in range(n-k-1,-1,-1): ## Right rotation
        arr[i+k] = arr[i]

    for i in range(len(temp)):
        arr[i] = temp[i]
    return arr
```

```python
arr = [1,2,3,4,5,6,7,8,9]
k = 2

right_rotate_k(arr,k)
```

# rotate by k places - optimized <br> - Done
arr = [1,2,3,4,5] k = 2  output = [4,5,1,2,3] <br>
Intuition - take k slice , move other elements, append in front Space Complexity O(n) <br>

<hr>
Optimized hack <br>
1. Reverse k :n <br>
2. Reverse 0:k <br>
3. Append and reverse full string

```python
def reverse_arr(sub_arr):
    return sub_arr[::-1]

arr = [1,2,3,4,5]

def k_rotation_arr(arr,k):
    n = len(arr)
    k = k % n

    arr[n-k:n] = reverse_arr(arr[n-k:n])
    arr[0:n-k] = reverse_arr(arr[0:n-k])
    arr[:] = reverse_arr(arr)

    return arr
```

# Max Consecutive Ones

arr = [1,1,0,1,1,1,1,0,1]  return 4

```python
arr = [1,1,2,1,1,1,1,1]

def get_max_consecutive_ones(arr):
    n = len(arr)
    count = 0
    max_count = 0
    i =0
    while i <n:
        if arr[i]==1:
            count = count+1
        else:
            if count > max_count:
                max_count = count
                count = 0
        i = i +1
    if count > max_count:
        max_count = count
    return max_count
```

# Brute force -  search for k-arr[i] in the subsequent array

```python
def check_2_sum(arr,k):
    n = len(arr)
    last_index = 0
    first_index = 0
    i =0
    while i<n-1:
        s = k-arr[i]
        for j in range(i+1,n):
            if arr[j] == s:
                last_index = j
                break
        if last_index!=0:
            first_index = i
            break
        i = i +1
    return first_index,last_index
    

```

```python
arr = [4,5,6,1,33]
k = 39
check_2_sum(arr,k)
```

## Optimal Solution using hashmap - O(n)

```python
hash_map = {}
arr = [4,5,6,1,33]
k = 38
first_index = 0
last_index = 0
n = len(arr)
for i in range(n):
    s = k - arr[i]
    if s in hash_map:
        first_index = i
        last_index = hash_map[s]
        break
    else:
        hash_map[arr[i]] = i
print(first_index,last_index)
```

# Maximum Subarray Sum - Kadane's Algorithm - Optimal Approach <hr>

nums = [1,2,-1,3,4,-3] <hr>

return 7 as [3,4]


```python
# Brute Force
def get_max_sub_sum(nums):

    total = 0
    maxi = float("-inf")
    n = len(nums)

    for i in range(n):
        total = 0
        for j in range(i,n):
            total = total+nums[j]
            if total>maxi:
                maxi = total
    return maxi

```

```python
nums = [1,2,-1,3,4,-3]

get_max_sub_sum(nums)
```

# Kadane's Algorithm - if the total is negative reset the total to zero

```python
nums = [1,2,-1,3,4,-3]
maxi = float("-inf")
total = 0
n = len(nums)

for i in range(n):
    total = total +nums[i]
    if total > maxi:
        maxi = total
    if total <0: ## Key step
        total = 0    
print(maxi)
```

# Stock Buy and Sell

## Brute Force - every possible combination

```python
prices = [7,2,1,5,6,4,8]

def get_buy_and_sell(arr):
    i = 0
    j = 0
    n = len(arr)
    sum_ = float("-inf")
    for i in range(0,n):
        print("Buy : ",arr[i],end= " ")
        for j in range(i+1,n):
            print("Sell : ",arr[j])
            diff = arr[j] - arr[i]
            if diff>sum_:
                sum_ = diff
    if sum_<0:
        return 0
    else:
        return sum_
```

```python
get_buy_and_sell(prices)
```

## optimized approach

```python
prices = [7,2,1,5,6,4,8]

def max_profit_sid_optimize(prices):
    max_profit = 0
    min_price = float("+inf")
    n = len(prices)
    for i in range(n):
        min_price = min(min_price,prices[i])
        max_profit = max(max_profit,prices[i]-min_price)
    return max_profit



    
```

```python
max_profit_sid_optimize(prices=prices)
```
