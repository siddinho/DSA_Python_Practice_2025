### Q3 - Check if array is sorted or not

```python
def check_array(arr):  ## Own Approach
    n = len(arr)
    asc = True

    if n <= 1:
        return True,asc
    elif n == 2:
        diff = arr[1]-arr[0]
        if diff>=0:
            return True,asc
        else:
            asc = False
            return True,asc
    else:
        diff = arr[1]-arr[0]
        if diff>=0:
            sort_p = True
            sort_n = False
        else:
            sort_n = True
            sort_p = False
            asc = False
        for i in range(n-1):
            diff = arr[i+1]-arr[i]
            if sort_p and diff<0:
                return False,False
            elif sort_n and diff>0:
                return False,False
            
    return True,asc

        
    
```

```python
arr = [2,1]
check_array(arr)
```

### Q4 Remove duplicates from a sorted array in place
input = [1,1,1,1,1,2,2,2,3,3,4] <br>
output = [1,2,3,4,-1,-1,-1,-1,-1]

```python
def remove_duplicates_sorted_arr(arr):
    if not arr:
        return 0

    write = 1

    for read in range(1, len(arr)):
        if arr[read] != arr[read - 1]:
            arr[write] = arr[read]
            write += 1

    return write  # length of unique array
```

```python
arr = [1,1,1,1,2,2,2,3,4,5]
k = remove_duplicates_sorted_arr(arr)
arr[0:k]
```

```python
def remove_dup_array(arr):
    n = len(arr)
    if n ==1:
        return 1
    i = 0
    j = 1
    while j <n:
        if arr[i]!=arr[j]:
            i = i +1
            arr[i],arr[j] = arr[j],arr[i]
        j= j +1
    return i+1 ## Unique elements
```

```python
arr  = [11,1,1,1,2,2,2,4,4,5,6]
k1 = remove_dup_array(arr)
arr[0:k1]
```

```python
def rotate_arr(arr):
    n = len(arr)
    temp = arr[n-1]
    for i in range(n-2,-1,-1):
        arr[i+1] =arr[i]
    arr[0] = temp
    return arr
```

# Rotate Right by k places

```python
def right_rotate_k(arr,k):
    n = len(arr)
    k = k%n ## Effective k 
    temp = arr[n-k:n]

    for i in range(n-k-1,-1,-1): ## Right rotation
        arr[i+k] = arr[i]

    for i in range(len(temp)):
        arr[i] = temp[i]
    return arr
```

```python
arr = [1,2,3,4,5,6,7,8,9]
k = 2

right_rotate_k(arr,k)
```

```python
arr = [1,2,0,0,3,4,0]
def z_end_brute_force(arr): # TC - O(n)  SC - O(n)
    n = len(arr)
    aux_list = []
    for i in arr:
        if i!=0:
            aux_list.append(i)
    print(aux_list)
    k = len(aux_list)
    arr[0:k] = aux_list
    arr[k:n] = [0]*(n-k)
    return arr
z_end_brute_force(arr)
```

```python
def z_end_optimal(arr):

    n = len(arr)
    i = 0
    while i <n:
        if arr[i]==0:
            break
        i = i +1
    if i ==n:
        return
    j = i+1
    while j<n:
        if arr[j]!=0:
            arr[i],arr[j] = arr[j],arr[i]
            i = i+1
        j = j +1


    
    


arr = [1,2,0,0,3,4,0]
z_end_optimal(arr)        

```

# rotate by k places - optimized <br> - Done
arr = [1,2,3,4,5] k = 2  output = [4,5,1,2,3] <br>
Intuition - take k slice , move other elements, append in front Space Complexity O(n) <br>

<hr>
Optimized hack <br>
1. Reverse k :n <br>
2. Reverse 0:k <br>
3. Append and reverse full string

```python
def reverse_arr(sub_arr):
    return sub_arr[::-1]

arr = [1,2,3,4,5]

def k_rotation_arr(arr,k):
    n = len(arr)
    k = k % n

    arr[n-k:n] = reverse_arr(arr[n-k:n])
    arr[0:n-k] = reverse_arr(arr[0:n-k])
    arr[:] = reverse_arr(arr)

    return arr
```

```python
k_rotation_arr(arr,3)
```

# Linear Search

```python
def l_search(arr,k):
    for i in range(len(arr)):
        if arr[i] ==k:
            return k
        
    return -1
```

```python
arr = [1,2,3,4,5,6]
k = 99
l_search(arr,k)
```

# Merge two sorted arrays without duplicates <br>

```python
arr = [1,1,2,2,3,4,5,6]
arr2 = [1,2,3,3,4,5,5] # output = [1,2,3,4,5,6]

m = len(arr)
n = len(arr2)
i = 0
j = 0
output_list = []
prev_value_pushed = float("-inf")
while i < m and j <n:
    if arr[i]==arr2[j] and arr[i]>prev_value_pushed:
        output_list.append(arr[i])
        prev_value_pushed = arr[i]
        i = i+1

    
    elif arr[i]<arr2[j]:
        val_to_be_pushed = arr[i]
        if val_to_be_pushed>prev_value_pushed:
            output_list.append(arr[i])
            prev_value_pushed = val_to_be_pushed
        i = i +1

    else:
        val_to_be_pushed = arr2[j]
        if val_to_be_pushed > prev_value_pushed:
            output_list.append(arr2[j])
            prev_value_pushed = val_to_be_pushed
        j = j+1


# Appending remaining elements
while i <m:
    if output_list[-1]<arr[i]:
        output_list.append(arr[i])
    i = i +1
while j<n:
    if output_list[-1]<arr2[j]:
        output_list.append(arr2[j])
    j = j +1

```

# Max Consecutive Ones

arr = [1,1,0,1,1,1,1,0,1]  return 4

```python
arr = [1,1,0,1,1,1,1,1]

def get_max_consecutive_ones(arr):
    n = len(arr)
    count = 0
    max_count = 0
    i =0
    while i <n:
        if arr[i]==1:
            count = count+1
        else:
            if count > max_count:
                max_count = count
                count = 0
        i = i +1
    if count > max_count:
        max_count = count
    return max_count
```

```python
get_max_consecutive_ones(arr)
```
