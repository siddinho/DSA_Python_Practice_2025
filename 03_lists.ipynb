### Q1 - Find the largest element in an array
sample_input = [3,5,7,8,1,2,1001]
return 1001

```python
## Own approach through Bubble Sort mechanism
arr = [3,4,5,1,2,11]
n = len(arr)
for i in range(0,n-2):
    if arr[i]>arr[i+1]:
        arr[i],arr[i+1] = arr[i+1],arr[i]
print(arr[n-1])
```

```python
## Max approach
arr = [3,4,5,1,2,11]
max_ = arr[0]
for i in range(len(arr)):
    if arr[i]>=max_:
        max_ = arr[i]
print(max_)
```

### Q3 - Check if array is sorted or not

```python
def check_array(arr):  ## Own Approach
    n = len(arr)
    asc = True

    if n <= 1:
        return True,asc
    elif n == 2:
        diff = arr[1]-arr[0]
        if diff>=0:
            return True,asc
        else:
            asc = False
            return True,asc
    else:
        diff = arr[1]-arr[0]
        if diff>=0:
            sort_p = True
            sort_n = False
        else:
            sort_n = True
            sort_p = False
            asc = False
        for i in range(n-1):
            diff = arr[i+1]-arr[i]
            if sort_p and diff<0:
                return False,False
            elif sort_n and diff>0:
                return False,False
            
    return True,asc

        
    
```

```python
arr = [2,1]
check_array(arr)
```

### Q4 Remove duplicates from a sorted array in place
input = [1,1,1,1,1,2,2,2,3,3,4] <br>
output = [1,2,3,4,-1,-1,-1,-1,-1]

```python
def remove_duplicates_sorted_arr(arr):
    if not arr:
        return 0

    write = 1

    for read in range(1, len(arr)):
        if arr[read] != arr[read - 1]:
            arr[write] = arr[read]
            write += 1

    return write  # length of unique array
```

```python
arr = [1,1,1,1,2,2,2,3,4,5]
k = remove_duplicates_sorted_arr(arr)
arr[0:k]
```

```python
def remove_dup_array(arr):
    n = len(arr)
    if n ==1:
        return 1
    i = 0
    j = 1
    while j <n:
        if arr[i]!=arr[j]:
            i = i +1
            arr[i],arr[j] = arr[j],arr[i]
        j= j +1
    return i+1 ## Unique elements
```

```python
arr  = [11,1,1,1,2,2,2,4,4,5,6]
k1 = remove_dup_array(arr)
arr[0:k1]
```

# Right Rotate an Array by 1 place

## With slicing

```python
arr = [1,2,3,4,5,6]

def right_rotate_slice(arr):
    n = len(arr)
    arr[:]= arr[n-1:]+arr[0:n-1] ##[:] for inplace rotation
    return arr

```

```python
right_rotate_slice(arr)
```

## Without slicing

```python
def rotate_arr(arr):
    n = len(arr)
    temp = arr[n-1]
    for i in range(n-2,-1,-1):
        arr[i+1] =arr[i]
    arr[0] = temp
    return arr
```

```python
rotate_arr([1,2,3,4,5,6])
```

# Rotate Right by k places

```python
def right_rotate_k(arr,k):
    n = len(arr)
    k = k%n ## Effective k 
    temp = arr[n-k:n]

    for i in range(n-k-1,-1,-1): ## Right rotation
        arr[i+k] = arr[i]

    for i in range(len(temp)):
        arr[i] = temp[i]
    return arr
```

```python
arr = [1,2,3,4,5,6,7,8,9]
k = 2

right_rotate_k(arr,k)
```
