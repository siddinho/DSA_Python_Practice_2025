### Q1 - Find the largest element in an array
sample_input = [3,5,7,8,1,2,1001]
return 1001

```python
## Own approach through Bubble Sort mechanism
arr = [3,4,5,1,2,11]
n = len(arr)
for i in range(0,n-2):
    if arr[i]>arr[i+1]:
        arr[i],arr[i+1] = arr[i+1],arr[i]
print(arr[n-1])
```

```python
## Max approach
arr = [3,4,5,1,2,11]
max_ = arr[0]
for i in range(len(arr)):
    if arr[i]>=max_:
        max_ = arr[i]
print(max_)
```

### Q2 - Find the second largest element in an array
sample_input = [3,5,7,8,1,2,1001]
return 8

```python
arr = [3,4,5,1,2,5]
arr = [5,4,3,2,1]
largest = float("-inf")
sl = float("-inf")
n = len(arr)

for i in range(n):
    if arr[i]>largest:
        sl = largest
        largest = arr[i]
    elif arr[i]>sl and arr[i]!=largest:
        sl = arr[i]
print(sl)
```
